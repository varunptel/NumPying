# RANDOM NUMBERS IN NUMPY

There are no true numbers in computer language as everything is just a set of instruction. So the random numbers generated by NumPy is just a deterministic algorithms. Therefore they are called pseudo-random numbers.

NumPy has built-in module numpy.random which is fast, vectorized and supports a lot of probability distributions.

The numpy.random module supplements the built-in Python random module with
functions for efficiently generating whole arrays of sample values from many kinds of probability distributions.


| Method | Description |
| :--- | :--- |
| `permutation` | Return a random permutation of a sequence, or return a permuted range. |
| `shuffle` | Randomly permute a sequence in place. |
| `uniform` | Draw samples from a uniform distribution (specify range). |
| `integers` | Draw random integers from a given low-to-high range. |
| `standard_normal` | Draw samples from a normal distribution with mean 0 and standard deviation 1. |
| `binomial` | Draw samples from a binomial distribution. |
| `normal` | Draw samples from a normal (Gaussian) distribution. |
| `beta` | Draw samples from a beta distribution. |
| `chisquare` | Draw samples from a chi-square distribution. |
| `gamma` | Draw samples from a gamma distribution. |
| `random` | Draw samples from a uniform [0, 1) distribution. |

## Generate Random Numbers:

```python
from numpy import random

x = random.randint(5) #integers
y = random.rand()   #random float

print(x)
print(y)
```

```python
# Generate random array

x = random.randint(50, size = (5)) #Integers 1-D. Change the size to (a,b) for 2D and so on.
y = random.rand(5) #1D float number random numbers array. (a,b) for 2D and so one.

print(x)
print(y)
```

```python
# To choose a random value from the array we can use choice() function from the random functions.

x = random.choice([35,69,45,55,56,78,3,55,2,66,0])

y = random.choice([35,69,45,55,56,78,3,55,2,66,0], size = (3,5)) #returns the array of size 3*5 from the array
print(x)
print(y)
```

```python
# Shuffling ararays

import numpy as np

arr = np.array([1,2,3,4,5,6])

np.random.shuffle(arr)
# Shuffle method make change to the original arrays.
print(arr)
```

```python
# Permutations of arrays

arr = np.array([1,2,3,4,5,6])

print(np.random.permutation(arr))  # Permutation function doesn't change the original array
print(arr)
```

### Modern Generator vs the Legacy interface

There are number of reason to switch from the legacy interface (the old np.random.rand(), np.random.randn(), np.random.seed(), etc) to modern generator(np.random.default_rng()):

1. It is Faster:
It uses modern, high-performance bit generator by default.

2. More flexible
- The Generator gives you way more control and features that the old API simply doesn’t have:

- You can pick the exact algorithm you want.

- Many more distributions and advanced sampling methods (e.g., multivariate_normal, hypergeometric, lognormal, etc.) that are either missing or less convenient in the legacy API.

```python
# Comparison between old and new random generators.
import numpy as np
import timeit

rng = np.random.default_rng()          # modern
legacy = np.random                    # old

# Generate 10 million standard normal numbers
print("Modern Generator:", timeit.timeit(lambda: rng.standard_normal(10_000_000), number=3))
print("Legacy np.random :", timeit.timeit(lambda: legacy.randn(10_000_000), number=3))
```

```python
import numpy as np

#creating a generator:
rng = np.random.default_rng(seed = 50) # seed = None : truly random each time

print(rng.random(5)) #Random float

print(rng.integers(low = 10, high = 50, size = (2,3)))  #Random Integer
```

## Random Data Distribution:

It means generating arrays of random numbers that follows specific probability distribution.

It is best practice to use it so that we can create data that behaves like real-world phenomena.

There are number of reasons to use it:
- Real data rarely follows uniform distribution
- Testing algorithms under realistic conditions
- Generate training data, initialize weights, run hypothesis tests and so on.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed = 40)
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.normal.html
data = rng.normal(loc=0, scale=1, size= 1000 * 1000)

plt.hist(data, bins = 50, density = True, alpha = 0.7)  #plot histogram with 50boxes
plt.show()
```
